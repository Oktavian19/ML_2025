# Naive Bayes

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/asset/ml/spam.csv', encoding='latin-1') # spesifiksi encoding diperlukan karena data tidak menggunakan UTF-8

df.head()
```

```python
df = df.iloc[:, :2]
df.columns = ['label', 'text']
df = df.rename(columns={'v1': 'Labels', 'v2': 'SMS'})

df['label_num'] = df['label'].map({'ham': 0, 'spam': 1})
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label_num'], test_size=0.2, random_state=42)

print(f"Training shape: {X_train.shape}")
print(f"Testing shape: {X_test.shape}")
```

## Fitur CountVectorizer

```python
count_vect = CountVectorizer(stop_words='english')

# transf train & test
X_train_count = count_vect.fit_transform(X_train)
X_test_count = count_vect.transform(X_test)

# Train MultinomialNB
nb_count = MultinomialNB()
nb_count.fit(X_train_count, y_train)

# Predict
y_pred_count = nb_count.predict(X_test_count)

# Evaluate
acc_count = accuracy_score(y_test, y_pred_count)
print(f"Accuracy (CountVectorizer): {acc_count}")
print("Classification Report (CountVectorizer):")
print(classification_report(y_test, y_pred_count))
```

## Fitur TF-IDF

```python
tf_idf = TfidfVectorizer(stop_words='english')

# transf train & test
X_train_tfidf = tf_idf.fit_transform(X_train)
X_test_tfidf = tf_idf.transform(X_test)

# Train MultinomialNB
nb_tfidf = MultinomialNB()
nb_tfidf.fit(X_train_tfidf, y_train)

# Predict
y_pred_tfidf = nb_tfidf.predict(X_test_tfidf)

# Evaluate
acc_tfidf = accuracy_score(y_test, y_pred_tfidf)
print(f"Accuracy (TF-IDF): {acc_tfidf}")
print("Classification Report (TF-IDF):")
print(classification_report(y_test, y_pred_tfidf))
```

```python
# Comparison
print("\n--- Comparison ---")
print(f"Accuracy CountVectorizer: {acc_count:.4f}")
print(f"Accuracy TF-IDF: {acc_tfidf:.4f}")
```

## Kesimpulan

Fitur CountVectorizer merupakan fitur yang lebih baik untuk mendeteksi spam pada dataset tersebut ketika dipasangkan dengna Multinominal Naive Bayes.

Meskipun TF-IDF menunjukkan nilai presisi di angka 100%, namun nilai recall jauh di bawahnya yaitu pada 75%. TF-IDF dinilai terlalu berhati-hati dalam mendeteksi pesan spam dan gagal mendeteksi seperempat pesan spam yang ada. Namun demikian, CountVectorizer dapat memberikan keselarasan pada skor presisi dan recall yaitu di angka 94% untuk nilai presisi dan 92% recall
