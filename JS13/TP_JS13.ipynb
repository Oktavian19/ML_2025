```python
import tensorflow as tf
import time
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import pandas as pd

# 1. Load dataset MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# Normalisasi data (0-255 → 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
def training_model(hidden_layers, activation_function):
  # 2. Bangun model JST
  model = Sequential()
  model.add(Flatten(input_shape=(28, 28)))

  for neuron in hidden_layers:
    model.add(Dense(neuron, activation=activation_function))

  model.add(Dense(10, activation='softmax'))

  # 3. Kompilasi model
  model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
  # 4. Latih model
  start_time = time.time()
  model.fit(X_train, y_train, epochs=5, batch_size=128, verbose=0)
  end_time = time.time()

  durasi = end_time - start_time
  # 5. Evaluasi model
  loss, acc = model.evaluate(X_test, y_test)
  print(f"Akurasi pada data uji: {acc:.4f}")
  return {
        'Struktur Layer': str(hidden_layers),
        'Aktivasi': activation_function,
        'Akurasi Test': acc,
        'Waktu Training (detik)': durasi
    }
```

```python
hasil_eksperimen = []

hasil_eksperimen.append(
    training_model([128, 64], 'relu')
)

# Ubah Jumlah Neuron
hasil_eksperimen.append(
    training_model([256, 128], 'relu')
)

# Tambah Hidden Layer
hasil_eksperimen.append(
    training_model([256, 128, 64], 'relu')
)

# Aktivasi Sigmoid
hasil_eksperimen.append(
    training_model([256, 128], 'sigmoid')
)
```

```python
df_hasil = pd.DataFrame(hasil_eksperimen)

print("\n" + "="*60)
print("TABEL PERBANDINGAN HASIL EKSPERIMEN MNIST")
print("="*60)
print(df_hasil)
```
