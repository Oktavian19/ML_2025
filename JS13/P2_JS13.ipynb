```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import tensorflow as tf
import numpy as np
import pandas as pd

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1, 1)
```

```python
# One-hot encoding
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
def model_training(hidden_1, hidden_2, activation):
  # Bangun model
  model = tf.keras.Sequential([
      tf.keras.layers.Dense(hidden_1, activation=activation, input_shape=(4,)),
      tf.keras.layers.Dense(hidden_2, activation=activation),
      tf.keras.layers.Dense(3, activation='softmax')
  ])

  # Kompilasi
  model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

  # Latih model
  model.fit(X_train, y_train, epochs=50, batch_size=8)

  # Evaluasi
  loss, acc = model.evaluate(X_test, y_test)
  print(f"Akurasi: {acc}")

  return {'Aktivasi': activation, 'Neuron L1': hidden_1, 'Neuron L2': hidden_2, 'Loss': loss, 'Akurasi': acc}
```

```python
hasil = []

r_base = model_training(10, 8, 'relu')
hasil.append(r_base)

r_neuron = model_training(64, 32, 'relu')
hasil.append(r_neuron)

r_sigmoid = model_training(10, 8, 'sigmoid')
hasil.append(r_sigmoid)
```

```python
print("\n" + "="*50)
print("TABEL PERBANDINGAN AKHIR")
print("="*50)
df_results = pd.DataFrame(hasil)
print(df_results)
```
